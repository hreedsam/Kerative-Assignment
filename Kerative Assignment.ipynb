# Task 1

```python
import csv
import os

with open(r"C:\Users\samar\Downloads\testing.csv", newline="") as f:
    hscodeIndex = -1
    productIndex = -1
    data = csv.reader(f)
    hscodeDiffs = {}
    for row in data:
        for keyCount in range(0,len(row)):
            if row[keyCount] == "HS CODE":
                hscodeIndex = keyCount
            if row[keyCount] == "PRODUCT":
                productIndex = keyCount
        if hscodeIndex != -1 and productIndex != -1:
            break
        
    for row in data:
        if row[hscodeIndex] == "" or row[productIndex] =="":
            continue
        if row[hscodeIndex] in hscodeDiffs:
            hscodeDiffs[row[hscodeIndex]].append(row[productIndex])
        else:
            hscodeDiffs[row[hscodeIndex]] = [row[productIndex]]


with open(r"C:\Users\samar\Downloads\solution.csv",'w',newline="") as f:
    csvPointer = csv.writer(f)
    csvPointer.writerow(["Sr. No.","HSCODE","Product Name"])
    cnt = 1
    for hscode in hscodeDiffs:
        csvPointer.writerow([cnt,hscode,hscodeDiffs[hscode][0]])
        for product in range(1,len(hscodeDiffs[hscode])):
            csvPointer.writerow(["",hscode,hscodeDiffs[hscode][product]])
        cnt += 1

```

# Task 2

```python
import csv
import os
hscodeFinalResult = {}


with open(r"C:\Users\samar\Downloads\testing.csv", newline="") as f:
    hscodeIndex = -1
    productIndex = -1
    quantityIndex = -1
    quantityUnitIndex = -1
    unitRateIndex = -1
    currencyIndex = -1
    totalUSDIndex = -1
    data = csv.reader(f)
    hscodeDiffs = {}
    for row in data:
        for keyCount in range(0,len(row)):
            if row[keyCount] == "HS CODE":
                hscodeIndex = keyCount
            if row[keyCount] == "PRODUCT":
                productIndex = keyCount
            if row[keyCount] == "QUANTITY":
                quantityIndex = keyCount
            if row[keyCount] == "UNIT":
                quantityUnitIndex = keyCount
            if row[keyCount] == "UNIT RATE":
                unitRateIndex = keyCount
            if row[keyCount] == "CURRENCY":
                currencyIndex = keyCount
            if row[keyCount] == "TOTAL USD":
                totalUSDIndex = keyCount
        if hscodeIndex != -1 and productIndex != -1:
            break
            
    for row in data:
        if row[hscodeIndex] == "" or row[productIndex] =="":
            continue
        if row[hscodeIndex] in hscodeFinalResult:
            if row[productIndex] not in hscodeFinalResult[row[hscodeIndex]]:
                temp = {
                    row[productIndex]: {
                        "quantity": row[quantityIndex],
                        "quantity unit": row[quantityUnitIndex],
                        "unit rate": row[unitRateIndex],
                        "currency": row[currencyIndex],
                        "total usd": row[totalUSDIndex]
                    }
                }
                hscodeFinalResult[row[hscodeIndex]].append(temp)

        else:
            temp = {
                row[productIndex]: {
                    "quantity": row[quantityIndex],
                    "quantity unit": row[quantityUnitIndex],
                    "unit rate": row[unitRateIndex],
                    "currency": row[currencyIndex],
                    "total usd": row[totalUSDIndex]
                }
            }
            hscodeFinalResult[row[hscodeIndex]] = [temp]
            
with open(r"C:\Users\samar\Downloads\solution2.csv",'w',newline="") as f:
    csvPointer = csv.writer(f)
    csvPointer.writerow(["Sr. No.","HSCODE","Product Name","Quantity","Unit","Unit Rate", "Currency","Total USD"])
    cnt = 1
    for hscode in hscodeFinalResult:
        sumQuantity = 0
        sumUnitRate = 0
        sumTotalUSD = 0
        for products in hscodeFinalResult[hscode]:
            for product in products:
                temp = [cnt,hscode,product,products[product]['quantity'],products[product]['quantity unit'],products[product]['unit rate'],products[product]['currency'],products[product]['total usd']]
                sumQuantity += float(products[product]['quantity'])
                sumUnitRate += float(products[product]['unit rate'])
                sumTotalUSD += float(products[product]['total usd'])
                csvPointer.writerow(temp)
                
        temp = ["","","Total",sumQuantity,"",sumUnitRate,"",sumTotalUSD]
        csvPointer.writerow(temp)
        cnt += 1
        

        
```
